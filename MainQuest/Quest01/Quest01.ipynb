#메인 퀘스트 1번: 은행 계좌 만들기#

```python
import random


class Account:  # Q1. Account 클래스

    account_count = 0  # Q2. 클래스 변수 - 계좌 객체의 개수

    def __init__(self, name, balance):  # 생성자 - 예금주, 초기 잔액만 입력
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

        self.name = name  # 예금주
        self.balance = balance  # 잔액
        self.bank = "SC은행"  # 은행 이름

        # 계좌 번호 (3자리-2자리-6자리 형태로 랜덤하게)
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)  # 1 -> '1' -> '001'
        num2 = str(num2).zfill(2)  # 1 -> '1' -> '01'
        num3 = str(num3).zfill(6)  # 1 -> '1' -> '0000001'
        self.account_number = num1 + '-' + num2 + '-' + num3  # 001-01-000001

        Account.account_count += 1  # 계좌 객체 개수 1 증가

    @classmethod
    def get_account_num(cls):  # Q3. 클래스 변수 출력 - 계좌 개수 출력
        print(cls.account_count)  # Account.account_count

    def deposit(self, amount):  # Q4. 입금 메서드
        if amount >= 1:  # 최소 1원 이상만 가능
            self.deposit_log.append(amount)
            self.balance += amount

            self.deposit_count += 1
            if self.deposit_count % 5 == 0:  # Q7. 이자 지급 - 입금 횟수가 5회가 될 때마다 (5, 10, 15회)
                print(f"{self.name}의 계좌에 {self.balance * 0.01}원의 이자가 지급되었습니다.")
                self.balance = (self.balance * 1.01)  # 잔고 기준 1% 이자 추가

        else:
            print("입금은 최소 1원 이상만 가능합니다.")

    def withdraw(self, amount):  # Q5. 출금 메서드
        if self.balance > amount:
            self.withdraw_log.append(amount)
            self.balance -= amount

        else:  # 계좌의 잔고 이상으로 출금할 수 없음
            print("출금은 계좌의 잔고 이상으로 출금할 수 없습니다.")

    def display_info(self):  # Q6. 정보 출력 메서드 - Account 인스턴스에 저장된 정보 출력
        print("은행이름: ", self.bank)
        print("예금주: ", self.name)
        print("계좌번호: ", self.account_number)
        print("잔고: ", self.balance)

    def withdraw_history(self):  # Q10-1. 출금 내역 출력
        for amount in self.withdraw_log:
            print(amount)

    def deposit_history(self):  # Q10-2. 입금 내역 출력
        for amount in self.deposit_log:
            print(amount)
```

```python
# Q8. 여러 객체 생성 - 3개 이상 인스턴스를 생성하여 리스트에 저장
data = []
c = Account("Coco", 3000000)
m = Account("Mimi", 500000)
s = Account("Sasha", 70000)
data.append(c)
data.append(m)
data.append(s)

print(f"생성된 계좌 개수: {Account.account_count}")

for h in data:
    h.display_info()
    print()

# Q9. 잔고가 100만원 이상인 고객의 정보만 출력
for h in data:
    if h.balance >= 1000000:
        h.display_info()
```

#메인 퀘스트 2번: 간단한 자동 사냥 RPG 만들기#

```python
import random

# Character 클래스 만들기
class Character:
    def __init__(self, name, level, physical, attack, defense):  # 이름, 레벨, 체력, 공격력, 방어력의 속성을 가짐
        self.name = name
        self.level = level
        self.physical = physical
        self.attack = attack
        self.defense = defense

# 인스턴스의 현재 체력이 0 이상인지 bool 값을 반환하는 is_alive 메서드 만들기
    def is_alive(self):
        if self.physical >= 0:
            return True
        else:
            return False

# 공격을 받았을 때, (받은 데미지 - 본인의 방어력)만큼 현재 체력이 감소하는 take_damage 메서드 만들기
    def take_damage(self, damage):
        if self.defense > damage:  # 본인의 방어력이 데미지보다 크다면 체력 감소하지 않음
            pass
        else:
            self.physical -= (damage - self.defense)
            print(f"{self.name}의 체력이 {damage - self.defense}만큼 감소했습니다.")

# 타켓에게 데미지를 입히는 attack_target 메서드 만들기
    def attack_target(self, target):
        damage = random.randint(1, self.attack)  # 데미지는 1부터 공격력 사이의 랜덤한 정수
        print(f"{self.name}이(가)) {target.name}에게 {damage}의 데미지를 입혔습니다.")
        target.take_damage(damage)
```

```python
# Character 클래스를 상속 받는 Player와 Monster 클래스를 만들기

# Player 클래스
class Player(Character):  # Character를 상속 받기
    def __init__(self, name):
        super().__init__(name, level=1, physical=100, attack=25, defense=5)

        # 레벨 1, 체력 100, 공격력 25, 방어력 5로 초기화하기
        #self.level = 1
        #self.physical = 100
        #self.attack = 25
        #self.defense = 5

        self.experience = 0  # 경험치 속성 추가
# Player 클래스는 경험치 속성을 추가로 가짐, 인수로 받은 정수 만큼 경험치를 획득하는 gain_experience 메서드 만들기
    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}의 경험치가 {exp}만큼 상승했습니다.")
        self.level_up()

# 현재 경험치가 50 이상이면 레벨을 1, 공격력을 10, 방어력을 5씩 올리는 level_up 메서드 만들기
    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name}의 레벨이 {self.level}이 되었습니다.")

# Monster 클래스
class Monster(Character):  # Character를 상속 받기
    def __init__(self, name, level):
        super().__init__(name, level, physical=100, attack=25, defense=5)

        # 몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화하기
        #self.level = level
        self.physical = random.randint(10, 30) * level  # 체력: 10~30 사이의 랜덤한 정수 * 레벨
        self.attack = random.randint(5, 20) * level  # 공격력: 5~20 사이의 랜덤한 정수 * 레벨
        self.defense = random.randint(1, 5) * level  # 방어력: 1~5 사이의 랜덤한 정수 * 레벨
```

```python
# battle 함수 만들기
def battle(player, monster):  # Player 인스턴스와 Monster 인스턴스를 인수로 받아
    print(f"{player.name}과(와) {monster.name}이(가) 전투를 시작했습니다.")

    while player.is_alive() and monster.is_alive():  # 둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고 받는 함수
        player.attack_target(monster)
        print(f"{monster.name}의 체력: {monster.physical}")
        print(f"{player.name}의 체력: {player.physical}")
        if monster.is_alive():
            monster.attack_target(player)
            print(f"{monster.name}의 체력: {monster.physical}")
            print(f"{player.name}의 체력: {player.physical}")

    if player.is_alive():  # 만약 Player 인스턴스가 살아 남았다면
        print(f"{player.name}이(가) 승리했습니다.")
        player.gain_experience(monster.level * 20)  # Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가
        player.level_up()  # Player의 레벨업 메서드를 호출하고
        print(f"{player.name}의 경험치: {player.experience}")  # Player의 경험치를 출력
        print("전투 승리!")  # '전투 승리!'를 출력
        print(f"{player.name}이(가) {monster.name}을(를) 이겼습니다.")
    else:
        print("전투 패배..")  # Player 인스턴스가 살아 남지 못했을 경우 '전투 패배..' 출력
        return False  # False 반환

    return True  # True 반환
```

```python
# main 함수 만들기
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}  # 몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기
    name = input("플레이어 이름을 입력하세요: ")  # 사용자로부터 이름을 입력받아 Player 인스턴스 생성하기
    player = Player(name)

    monsters = []
    for name, level in monster_dict.items():  # 몬스터 딕셔너리로부터 Monster 인스턴스 생성하기
        monster = Monster(name, level)
        monsters.append(monster)

    for monster in monsters:  # player는 생성된 몬스터 3마리(슬라임, 고블린, 오크)와 모두 전투해야함
        battle(player, monster)  # 생성된 Monster 인스턴스와 Player 인스턴스가 battle 함수를 통해 전투
        if not player.is_alive():
            print("게임 오버")  # 전투 도중에 Player가 사망하면 이후 전투를 진행하지 않고 ‘게임오버’ 출력하고 종료
            break

if __name__ == "__main__":
    main()



```
